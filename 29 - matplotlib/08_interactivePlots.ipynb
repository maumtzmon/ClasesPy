# Plots Interactivos
### Una manera de expresar el comportamiento o tendencia de un fenomeno es a partir de la evolucion de los datos de un experimento.
### Una grafica puede representar dicho comportamiento con condiciones iniciales. Esas condiciones, constantes al inicio, podemos variarlas para comparar el mismo fenomeno de diferentes maneras dentro de una sola grafica.

```python
import ipywidgets
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.uniform(0,5, size = 100)
noise = np.random.normal(size=100)
```

```python
def plot_fct(w=1): #la manera de hacer interactiva la grafica es definir una variable en el parametro de la funcion....
    y=2*X+w*noise
    plt.scatter(X,y)
```

```python
ipywidgets.interact(plot_fct,w=(0,5,0.5)) # La variable con la que deseamos interactuar en la grafica debe ingresarse como parametro y establecer un rango 
                                          # el parametro w varia desde 0 hasta 5 en pasos de 0.5
```

```python
from sklearn.datasets import make_moons
```

```python
moons =make_moons(n_samples=100,noise=0.09)
X,y = moons[0], moons[1]
plt.scatter(X[:,0],X[:,1],c=y)            # La grafica contiene valores que podemos utilizar como variable de interaccion, por ejemplo el numero de muestras y el ruido
```

### Definamos el codigo anterior como una funcion y como "parametros variables o interactivos" el ruido y numero de muestras

```python
def plot_moons(samples=200, noise=0):
    moons=make_moons(n_samples=samples, noise=noise, random_state=50)
    X,y=moons[0], moons[1]
    plt.scatter(X[:,0], X[:,1],c=y)
```

```python
ipywidgets.interact(plot_moons,samples=[200,500,1000],noise=(0,0.5,0.0005))
```
